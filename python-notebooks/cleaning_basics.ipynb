## Grouping
Since we are comparing white with non-white students, we will group by ethnicities

```python
non_white_students = {'Non-white Students': ['American Indian/Alaskan Native', 'Asian', 'Black/African American', 'Hispanic/Latino of any race(s)', 'Native Hawaiian/Other Pacific Islander', 'Two or More Races'], 'White Students': ['White']}
df['Non-white Students'] = df[non_white_students['Non-white Students']].sum(axis=1)
df['White Students'] = df[non_white_students['White Students']].sum(axis=1)
```

```python
df.head()
```

## Remove rows

```python
# Remove rows where 'GradeLevel' is 'All Grades' to better perform calculations and averages
df = df[df['GradeLevel'] != 'All Grades']
```

```python
df.head()
```

```python
# Remove rows where number of students = 0
df = df[df['White Students'] + df['Non-white Students'] != 0]
```

## Combine columns with the same district
Districts will combine all their data and produce a cohesive data center for 'All Grades'

```python
grouped_df = df.groupby('DistrictName')[['White Students', 'Non-white Students', 'Low Income', 'Non-Low Income', 'Highly Capable', 'Non-Highly Capable', 'English Language Learners',
                                         'Non-English Language Learners', 'Students with Disabilities', 'Students without Disabilities',
                                         'Homeless', 'Non-Homeless', 'Migrant', 'Non-Migrant']].sum().reset_index()
```

```python
grouped_df.head()
```

# Calculations
Find districts with highest number of difference in student population

```python
# Find districts with highest number of difference in student population
grouped_df['Difference'] = (grouped_df['White Students'] - grouped_df['Non-white Students']) / (grouped_df['White Students'] + grouped_df['Non-white Students'])
sorted_df = grouped_df.sort_values(by='Difference', ascending=False)
```

```python
sorted_df.head()
```

```python
sorted_df.tail()
```

```python
majority_white = sorted_df[sorted_df['Difference'] > 0.5].copy()
majority_non_white = sorted_df[sorted_df['Difference'] < -0.5].copy()
```

# Visualizations
Showcase results

```python
# Showcase results by comparing district charts with high number of white students vs low number and vice versa
# See the differences in majorite white students vs non-white students districts
import matplotlib.pyplot as plt

# Compare average values
features = ['Low Income', 'Homeless', 'Migrant', 'Highly Capable', 'Non-Highly Capable']

# Mean values per group
white_avg = majority_white[features].mean()
non_white_avg = majority_non_white[features].mean()

# Bar chart comparison
x = range(len(features))
plt.figure(figsize=(10,6))
plt.bar(x, white_avg, width=0.4, label='Majority White', align='center')
plt.bar([i + 0.4 for i in x], non_white_avg, width=0.4, label='Majority Non-White', align='center')

plt.xticks([i + 0.2 for i in x], features)
plt.ylabel('Average Student Count')
plt.title('Comparison of Risk Factors in Majority White vs Non-White Districts')
plt.legend()
plt.show()
```
